```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile
from scipy.fft import fft, fftfreq

def fftplot(data,rate):
    N = data.shape[0]
    T = 1/rate
    yf = fft(data)
    xf = fftfreq(N, T)[:N//2]
    plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
    plt.grid()
    plt.show()

#load data 
#rate, data = wavfile.read("p232_013.wav") 
#print(rate)
#fftplot(data,rate)
#wavfile.write("p232_013-1.wav", rate, (data/2).astype(np.int16));
#plt.plot(data[0:1000])
#plt.show()

samplerate = 44100; fs = 200
t = np.linspace(0., 1., samplerate)
amplitude = np.iinfo(np.int16).max
data = amplitude * np.sin(2. * np.pi * fs * t)
fftplot(data, samplerate)
#plt.plot(data)
#plt.show()
```
